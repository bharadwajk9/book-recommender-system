```python
ratings_counts = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
ratings_counts = ratings_counts.sort_values('bookRating' ,ascending = False).head()
```

```python
ratings_counts
```

```python
most_rated_books_summary = pd.merge(ratings_counts, books, on = 'ISBN')
most_rated_books_summary
```

```python
average_ratings = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].mean())
average_ratings['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
average_ratings = average_ratings.sort_values('ratingCount', ascending =False)
```

```python
average_ratings.head()
```

```python
counts1 = ratings.userID.value_counts()

#1149780
ratings = ratings[ratings['userID'].isin(counts1[counts1 >= 200].index)]
ratings.shape
#527556

counts = ratings.bookRating.value_counts()
ratings = ratings[ratings['bookRating'].isin(counts[counts >= 100].index)]
ratings.shape
```

```python
ratings_pivot = ratings.pivot(index = 'userID', columns = 'ISBN').bookRating
ratings_pivot.head()
```

```python
bones_ratings = ratings_pivot['0316666343']
similar_to_bones = ratings_pivot.corrwith(bones_ratings)
corr_bones = pd.DataFrame(similar_to_bones, columns = ['pearsonR'])
corr_bones.dropna(inplace = True)
corr_summary = corr_bones.join(average_ratings['ratingCount'])
corr_summary_head = corr_summary[corr_summary['ratingCount'] > 300].sort_values('pearsonR', ascending =False).head(10)
corr_summary_head
```

```python
books_corr_to_bones = pd.DataFrame(['0312291639', '0316601950', '0446610038', '0446672211', '0385265700', '0345342968', '0060930535', '0375707972', '0684872153'], 
                                  index=np.arange(9), columns=['ISBN'])
corr_books = pd.merge(books_corr_to_bones, books, on = 'ISBN')
corr_books
```
